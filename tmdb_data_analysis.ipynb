# Data Preprocessing

we can see that the data contains a lot of columns that are a list of dictionaries that's why we are going to extract the important info from each of these list into a singl column

```python
def extract_name_to_list(lst):
    return '|'.join([dic['name'] for dic in eval(lst)])
data['genres'] = data['genres'].apply(extract_name_to_list)
data['keywords'] = data['keywords'].apply(extract_name_to_list)
data['production_companies'] = data['production_companies'].apply(extract_name_to_list)
data['production_countries'] = data['production_countries'].apply(extract_name_to_list)
data['spoken_languages'] = data['spoken_languages'].apply(extract_name_to_list)
data.head()
```

Now let us Look if there are missing data in our column and their percentage from the total number of rows

```python
pd.concat([(data.isnull().sum() *100 / data.shape[0]).sort_values(ascending = False)
           .rename('count_pcg') 
           ,data.isnull().sum().sort_values(ascending = False)
           .rename('counts') ], axis = 1)
```

from the above we can see that most of our columns have no missing data except of "homepage" column which has more that 60% missing data. In addition to "tagline" column that have 17% missing data. For the purpose of our analysis both columns can be dropped safely as they don't contain information relative to our analysis. Regarding the other 3 columns, I will impute the "runtime" with its mean and drop rows contains null values for the other 2 columns. 

```python
data.drop(['homepage', 'tagline'], axis = 1, inplace = True)
data['runtime'].fillna(data['runtime'].mean(), inplace = True)
data.dropna(inplace=True)
```

### Q7: what are top 10 highly rated movies?

```python
(data.sort_values('vote_average', ascending = False)[['title', 'vote_average', 'vote_count']]
     .head(10)
)
```

As we can see that the top 10 rated movies contain movies that have vote average of 10 however the vote_count for them is too few to be representative for the top vote_average. In the below figure, there is a lot of data points that have very high rating however number of people who have given this rating are to few.

```python
plt.figure(figsize=(10,8))
plt.title('Relationship between the average rating and vote counts')
_ = sns.scatterplot('vote_count','vote_average', data = data)
```

I am going to use Gaussian mixture model to cluster data point into 2 cluster to identify out of pattern data (outlier data) in 2 dimensions.

```python
from sklearn.mixture import GaussianMixture
model = GaussianMixture(n_components=2, random_state=100)
vote_count_outliers_flag = model.fit_predict(data[['vote_count','vote_average']])
vote_count_outliers_flag
```

```python
plt.figure(figsize=(10,8))
plt.title('Relationship between the average rating and vote counts After Clustering')
sns.scatterplot(
        x = 'vote_count'
        ,y = 'vote_average'
        ,hue = ['not_outlier' if x == 1 else 'outlier' for x in vote_count_outliers_flag  ]
        , data = data)
```

```python
data['vote_count_outlier_flag'] = vote_count_outliers_flag
```

```python
(data[vote_count_outliers_flag == 1]
     .sort_values('vote_average', ascending = False)[['title', 'vote_average', 'vote_count']]
     .head(10)
)
```

In this list we were able to obtain top 10 rating with very resonable number of vote count for each movie so that it can be represenative for the provided movies rating.
